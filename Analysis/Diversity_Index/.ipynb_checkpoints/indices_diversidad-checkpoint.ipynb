```python
#Script para comparar los tres indices (D , 1/D y H')


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def simpson_diversity_index(counts):
    total = sum(counts)
    if total == 0:
        return 0
    return sum((n / total) ** 2 for n in counts)

def simpson_reciprocal_index(counts):
    total = sum(counts)
    if total == 0:
        return 0
    D = sum((n / total) ** 2 for n in counts)
    return 1 / D if D > 0 else 0  # Evita la división por cero

def shannon_diversity_index(counts):
    total = sum(counts)
    if total == 0:
        return 0
    proportions = counts / total
    H = -sum(p * np.log(p) for p in proportions if p > 0)  # Evitar log(0)
    return H

def calculate_alpha_diversity(csv_file, community_col, species_col, output_file=None):
    df = pd.read_csv(csv_file)
    
    if community_col not in df.columns or species_col not in df.columns:
        raise ValueError(f"El archivo CSV debe contener las columnas '{community_col}' y '{species_col}'.")
    
    community_counts = df.groupby([community_col, species_col]).size().unstack(fill_value=0)
    
    diversity_results = pd.DataFrame({
        community_col: community_counts.index,
        'Simpson_Diversity_Index': community_counts.apply(simpson_diversity_index, axis=1),
        'Simpson_Reciprocal_Index': community_counts.apply(simpson_reciprocal_index, axis=1),
        'Shannon_Diversity_Index': community_counts.apply(shannon_diversity_index, axis=1)
    })
    
    if output_file:
        diversity_results.to_csv(output_file, index=False)
        print(f"Resultados guardados en {output_file}")
    
    return diversity_results

def plot_diversity_indices(results_df, community_col, output_prefix="diversidad"):
    indices = [
        ('Simpson_Diversity_Index', "Índice de Diversidad de Simpson (D)"),
        ('Simpson_Reciprocal_Index', "Índice de Diversidad de Simpson Recíproco (1/D)"),
        ('Shannon_Diversity_Index', "Índice de Diversidad de Shannon (H')")
    ]
    
    for col, title in indices:
        plt.figure(figsize=(10, 5))
        plt.bar(results_df[community_col], results_df[col], color='skyblue', edgecolor='black')
        plt.xlabel("Comunidades")
        plt.ylabel(title)
        plt.title(f"Diversidad Alfa - {title}")
        plt.xticks(rotation=45, ha='right')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        
        output_plot = f"{output_prefix}_{col.lower()}.png"
        plt.savefig(output_plot, dpi=600, bbox_inches='tight')
        print(f"Gráfico guardado en {output_plot}")
        plt.show()

# --- EJEMPLO DE USO ---
csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"  # Ruta del archivo CSV
community_col = "Sample"  # Nombre de la columna que define la comunidad
species_col = "phenotype_key"  # Nombre de la columna que define las especies
output_file = "diversidad_indices.csv"

diversity_results = calculate_alpha_diversity(csv_file, community_col, species_col, output_file)
plot_diversity_indices(diversity_results, community_col)
```

### Índice de Diversidad de Simpson (D)

El índice de diversidad de Simpson se calcula con la siguiente fórmula:


$D = \sum_{i=1}^{S} \left(\frac{n_i}{N}\right)^2$

donde:

- $D$ = Índice de diversidad de Simpson.
- $S$ = Número total de especies en la comunidad.
- $n_i$ = Número de individuos de la especie $i$.
- $N$ = Número total de individuos en la comunidad.

### Interpretación:
- **Valores cercanos a 1** → **Menor diversidad**, dominan pocas especies.
- **Valores cercanos a 0** → **Mayor diversidad**, distribución más equitativa de las especies.


```python
import pandas as pd
import matplotlib.pyplot as plt

def simpson_diversity_index(counts):
    """
    Calcula el índice de diversidad de Simpson (D).
    
    Parámetros:
    counts (list): Lista de abundancias de cada especie en la comunidad.

    Retorna:
    float: Índice de diversidad de Simpson.
    """
    total = sum(counts)
    if total == 0:
        return 0
    return sum((n / total) ** 2 for n in counts)

def calculate_alpha_diversity(csv_file, output_file=None):
    """
    Calcula la diversidad alfa (índice de Simpson D) para cada comunidad.

    Parámetros:
    csv_file (str): Ruta al archivo CSV con columnas 'ROIs' y 'phenotype_key'.
    output_file (str, opcional): Ruta para guardar los resultados en un CSV.

    Retorna:
    pd.DataFrame: DataFrame con los índices de diversidad de Simpson por comunidad.
    """
    # Cargar datos
    df = pd.read_csv(csv_file)

    # Verificar si las columnas requeridas están en el archivo
    if 'Sample' not in df.columns or 'phenotype_key' not in df.columns:
        raise ValueError("El archivo CSV debe contener las columnas 'ROIs' y 'phenotype_key'.")

    # Contar individuos por especie en cada comunidad
    community_counts = df.groupby(['Sample', 'phenotype_key']).size().unstack(fill_value=0)

    # Calcular índice de Simpson para cada comunidad
    diversity_scores = community_counts.apply(lambda row: simpson_diversity_index(row), axis=1)

    # Crear DataFrame de resultados
    results = pd.DataFrame({'Sample': diversity_scores.index, 'Simpson_Diversity_Index': diversity_scores.values})

    # Guardar resultados si se especifica un archivo de salida
    if output_file:
        results.to_csv(output_file, index=False)
        print(f"Resultados guardados en {output_file}")

    return results

def plot_alpha_diversity(results_df, output_plot="diversidad_simpson.png"):
    """
    Genera y guarda un gráfico de barras del índice de diversidad de Simpson.

    Parámetros:
    results_df (pd.DataFrame): DataFrame con columnas 'ROIs' y 'Simpson_Diversity_Index'.
    output_plot (str, opcional): Nombre del archivo donde se guardará la gráfica.
    """
    plt.figure(figsize=(10, 5))
    plt.bar(results_df['Sample'], results_df['Simpson_Diversity_Index'], color='skyblue', edgecolor='black')
    plt.xlabel("Comunidades (Sample)")
    plt.ylabel("Índice de Diversidad de Simpson")
    plt.title("Diversidad Alfa - Índice de Simpson (D)")
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 1)  # Ajustar la escala del índice de 0 a 1
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Guardar la gráfica
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")

    plt.show()

# --- EJEMPLO DE USO ---

csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"  # Reemplaza con la ruta de tu archivo CSV
output_file = "diversidad_simpson.csv"  # Opcional: Archivo donde se guardan los datos
output_plot = "diversidad_simpson.png"  # Nombre del archivo para la gráfica

# Calcular diversidad
resultados = calculate_alpha_diversity(csv_file, output_file)

# Graficar los resultados
plot_alpha_diversity(resultados, output_plot)
```

### Índice Recíproco de Simpson (1/D)

El índice recíproco de Simpson se calcula con la siguiente fórmula:

$1/D = \frac{1}{\sum_{i=1}^{S} \left(\frac{n_i}{N}\right)^2}$

donde:

- $1/D$ = Índice recíproco de Simpson.
- $S$ = Número total de especies en la comunidad.
- $n_i$ = Número de individuos de la especie $i$.
- $N$ = Número total de individuos en la comunidad.

### Interpretación:
- **Valores bajos** ($1/D \approx 1$) → Baja diversidad, pocas especies dominantes.
- **Valores altos** ($1/D > 3$) → Alta diversidad, más especies distribuidas equitativamente.


-------------------------------------

```python
import pandas as pd
import matplotlib.pyplot as plt

def simpson_reciprocal_index(counts):
    """
    Calcula el índice de diversidad de Simpson recíproco (1/D).
    
    Parámetros:
    counts (list): Lista de abundancias de cada especie en la comunidad.

    Retorna:
    float: Índice de diversidad de Simpson recíproco (1/D).
    """
    total = sum(counts)
    if total == 0:
        return 0
    D = sum((n / total) ** 2 for n in counts)
    return 1 / D if D > 0 else 0  # Evita la división por cero

def calculate_alpha_diversity(csv_file, output_file=None):
    """
    Calcula la diversidad alfa (índice de Simpson recíproco 1/D) para cada comunidad.

    Parámetros:
    csv_file (str): Ruta al archivo CSV con columnas 'ROIs' y 'phenotype_key'.
    output_file (str, opcional): Ruta para guardar los resultados en un CSV.

    Retorna:
    pd.DataFrame: DataFrame con los índices de diversidad de Simpson recíproco por comunidad.
    """
    # Cargar datos
    df = pd.read_csv(csv_file)

    # Verificar si las columnas requeridas están en el archivo
    if 'Sample' not in df.columns or 'phenotype_key' not in df.columns:
        raise ValueError("El archivo CSV debe contener las columnas 'Sample' y 'phenotype_key'.")

    # Contar individuos por especie en cada comunidad
    community_counts = df.groupby(['Sample', 'phenotype_key']).size().unstack(fill_value=0)

    # Calcular índice de Simpson recíproco para cada comunidad
    diversity_scores = community_counts.apply(lambda row: simpson_reciprocal_index(row), axis=1)

    # Crear DataFrame de resultados
    results = pd.DataFrame({'Sample': diversity_scores.index, 'Simpson_Reciprocal_Index': diversity_scores.values})

    # Guardar resultados si se especifica un archivo de salida
    if output_file:
        results.to_csv(output_file, index=False)
        print(f"Resultados guardados en {output_file}")

    return results

def plot_alpha_diversity(results_df, output_plot="diversidad_simpson.png"):
    """
    Genera y guarda un gráfico de barras del índice de diversidad de Simpson recíproco (1/D).

    Parámetros:
    results_df (pd.DataFrame): DataFrame con columnas 'ROIs' y 'Simpson_Reciprocal_Index'.
    output_plot (str, opcional): Nombre del archivo donde se guardará la gráfica.
    """
    plt.figure(figsize=(10, 5))
    plt.bar(results_df['Sample'], results_df['Simpson_Reciprocal_Index'], color='skyblue', edgecolor='black')
    plt.xlabel("Comunidades (Sample)")
    plt.ylabel("Índice de Diversidad de Simpson Recíproco (1/D)")
    plt.title("Diversidad Alfa - Índice de Simpson Recíproco (1/D)")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Guardar la gráfica
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")

    plt.show()

# --- EJEMPLO DE USO ---

csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"  # Reemplaza con la ruta de tu archivo CSV
output_file = "diversidad_simpson_reciproco.csv"  # Opcional: Archivo donde se guardan los datos
output_plot = "diversidad_simpson_reciproco.png"  # Nombre del archivo para la gráfica

# Calcular diversidad
resultados = calculate_alpha_diversity(csv_file, output_file)

# Graficar los resultados
plot_alpha_diversity(resultados, output_plot)
```

### Índice de Diversidad de Shannon (H')

El índice de diversidad de Shannon se calcula con la siguiente fórmula:

$H' = -\sum_{i=1}^{S} p_i \ln(p_i)$

donde:

- $H'$ = Índice de diversidad de Shannon.
- $S$ = Número total de especies en la comunidad.
- $p_i$ = Proporción de individuos de la especie $i$, calculada como:

$p_i = \frac{n_i}{N}$

  - $n_i$ = Número de individuos de la especie $i$.
  - $N$ = Número total de individuos en la comunidad.

### Interpretación:
- **Valores bajos** ($H' \approx 0$) → Poca diversidad (una especie dominante).
- **Valores altos** ($H' > 3$) → Alta diversidad con especies bien distribuidas.


-------------------------------------

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def shannon_diversity_index(counts):
    """
    Calcula el índice de diversidad de Shannon (H').

    Parámetros:
    counts (list): Lista de abundancias de cada especie en la comunidad.

    Retorna:
    float: Índice de diversidad de Shannon.
    """
    total = sum(counts)
    if total == 0:
        return 0
    proportions = counts / total
    H = -sum(p * np.log(p) for p in proportions if p > 0)  # Evitar log(0)
    return H

def calculate_alpha_diversity(csv_file, output_file=None):
    """
    Calcula la diversidad alfa (índice de Shannon) para cada comunidad.

    Parámetros:
    csv_file (str): Ruta al archivo CSV con columnas 'ROIs' y 'phenotype'.
    output_file (str, opcional): Ruta para guardar los resultados en un CSV.

    Retorna:
    pd.DataFrame: DataFrame con los índices de diversidad de Shannon por comunidad.
    """
    # Cargar datos
    df = pd.read_csv(csv_file)

    # Verificar si las columnas requeridas están en el archivo
    if 'Sample' not in df.columns or 'phenotype_key' not in df.columns:
        raise ValueError("El archivo CSV debe contener las columnas 'Sample' y 'phenotype_key'.")

    # Contar individuos por especie en cada comunidad
    community_counts = df.groupby(['Sample', 'phenotype_key']).size().unstack(fill_value=0)

    # Calcular índice de Shannon para cada comunidad
    diversity_scores = community_counts.apply(lambda row: shannon_diversity_index(row), axis=1)

    # Crear DataFrame de resultados
    results = pd.DataFrame({'Sample': diversity_scores.index, 'Shannon_Diversity_Index': diversity_scores.values})

    # Guardar resultados si se especifica un archivo de salida
    if output_file:
        results.to_csv(output_file, index=False)
        print(f"Resultados guardados en {output_file}")

    return results

def plot_alpha_diversity(results_df, output_plot="diversidad_shannon.png"):
    """
    Genera y guarda un gráfico de barras del índice de diversidad de Shannon.

    Parámetros:
    results_df (pd.DataFrame): DataFrame con columnas 'ROIs' y 'Shannon_Diversity_Index'.
    output_plot (str, opcional): Nombre del archivo donde se guardará la gráfica.
    """
    plt.figure(figsize=(10, 5))
    plt.bar(results_df['Sample'], results_df['Shannon_Diversity_Index'], color='skyblue', edgecolor='black')
    plt.xlabel("Comunidades (Sample)")
    plt.ylabel("Índice de Diversidad de Shannon (H')")
    plt.title("Diversidad Alfa - Índice de Shannon (H')")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Guardar la gráfica
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")

    plt.show()

# --- EJEMPLO DE USO ---

csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"  # Ruta del archivo CSV
output_file = "diversidad_shannon.csv"  # Archivo donde se guardan los datos
output_plot = "diversidad_shannon.png"  # Nombre del archivo para la gráfica

# Calcular diversidad
resultados = calculate_alpha_diversity(csv_file, output_file)

# Graficar los resultados
plot_alpha_diversity(resultados, output_plot)
```

### Indice inverso de simpson 1/D - points

```python
sample_to_analyze = "FAAGSCC_9"  # Comunidad a analizar
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def simpson_reciprocal_index(counts):
    """
    Calcula el índice de diversidad de Simpson recíproco (1/D).
    """
    total = sum(counts)
    if total == 0:
        return 0
    D = sum((n / total) ** 2 for n in counts)
    return 1 / D if D > 0 else 0

def calculate_spatial_diversity(df, community_col, species_col, sample, grid_size=450):
    """
    Calcula la diversidad alfa espacialmente dividiendo el área en una cuadrícula.
    """
    df_sample = df[df[community_col] == sample].copy()
    
    # Definir los límites del espacio
    x_min, x_max = df_sample['X_centroid'].min(), df_sample['X_centroid'].max()
    y_min, y_max = df_sample['Y_centroid'].min(), df_sample['Y_centroid'].max()
    
    # Crear la cuadrícula
    x_bins = np.arange(x_min, x_max, grid_size)
    y_bins = np.arange(y_min, y_max, grid_size)
    
    # Asignar celdas a la cuadrícula
    df_sample['X_bin'] = np.digitize(df_sample['X_centroid'], x_bins)
    df_sample['Y_bin'] = np.digitize(df_sample['Y_centroid'], y_bins)
    
    # Agrupar por celda y calcular diversidad
    diversity_results = df_sample.groupby(['X_bin', 'Y_bin'])[species_col].apply(lambda x: simpson_reciprocal_index(x.value_counts())).reset_index()
    diversity_results.columns = ['X_bin', 'Y_bin', 'Simpson_Reciprocal_Index']
    
    # Convertir celdas a coordenadas espaciales
    diversity_results['X_coord'] = x_bins[diversity_results['X_bin'] - 1] + grid_size / 2
    diversity_results['Y_coord'] = y_bins[diversity_results['Y_bin'] - 1] + grid_size / 2
    
    return diversity_results

def plot_spatial_diversity(diversity_df, sample, output_plot="diversidad_espacial.png"):
    """
    Genera y guarda un mapa de calor de la diversidad alfa espacial dentro de una comunidad.
    """
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(diversity_df['X_coord'], diversity_df['Y_coord'], 
                          c=diversity_df['Simpson_Reciprocal_Index'], cmap='viridis', 
                          edgecolors='black', s=1, alpha=0.8)
    
    plt.colorbar(scatter, label="Índice de Diversidad 1/D")
    plt.xlabel("Posición X")
    plt.ylabel("Posición Y")
    plt.title(f"Diversidad Alfa Espacial en la Comunidad {sample}")
    
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")
    
    plt.show()

# --- CONFIGURACIÓN DE PARÁMETROS ---
csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"  # Ruta del archivo CSV
#sample_to_analyze = "AGSCC_1"  # Comunidad a analizar
output_plot = f"simpson_points_{sample_to_analyze}.png"
community_col = "Sample"  # Nombre de la columna con la comunidad
species_col = "phenotype_key"  # Nombre de la columna con el nombre de especies

df = pd.read_csv(csv_file)
diversity_results = calculate_spatial_diversity(df, community_col, species_col, sample_to_analyze, grid_size=450)
plot_spatial_diversity(diversity_results, sample_to_analyze, output_plot)
```

### Indice inverso de simpson 1/D - Heatmap
-------------------------------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def simpson_reciprocal_index(counts):
    """
    Calcula el índice de diversidad de Simpson recíproco (1/D).
    """
    total = sum(counts)
    if total == 0:
        return 0
    D = sum((n / total) ** 2 for n in counts)
    return 1 / D if D > 0 else 0

def calculate_spatial_diversity(df, community_col, species_col, sample, grid_size=100):
    """
    Calcula la diversidad alfa espacialmente dividiendo el área en una cuadrícula.
    """
    df_sample = df[df[community_col] == sample].copy()

    # Definir los límites del espacio
    x_min, x_max = df_sample['X_centroid'].min(), df_sample['X_centroid'].max()
    y_min, y_max = df_sample['Y_centroid'].min(), df_sample['Y_centroid'].max()

    # Crear la cuadrícula
    x_bins = np.arange(x_min, x_max, grid_size)
    y_bins = np.arange(y_min, y_max, grid_size)

    # Asignar cada punto a una celda de la cuadrícula
    df_sample['X_bin'] = np.digitize(df_sample['X_centroid'], x_bins)
    df_sample['Y_bin'] = np.digitize(df_sample['Y_centroid'], y_bins)

    # Agrupar por celda y calcular diversidad
    diversity_results = df_sample.groupby(['X_bin', 'Y_bin'])[species_col].apply(lambda x: simpson_reciprocal_index(x.value_counts())).reset_index()
    diversity_results.columns = ['X_bin', 'Y_bin', 'Simpson_Reciprocal_Index']

    return diversity_results, x_bins, y_bins

def plot_spatial_diversity(diversity_df, x_bins, y_bins, sample, output_plot="diversidad_espacial.png"):
    """
    Genera y guarda un mapa de calor de la diversidad alfa espacial dentro de una comunidad.
    """
    heatmap_data = np.full((len(y_bins), len(x_bins)), np.nan)

    for _, row in diversity_df.iterrows():
        heatmap_data[int(row['Y_bin']) - 1, int(row['X_bin']) - 1] = row['Simpson_Reciprocal_Index']

    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data, cmap='viridis', cbar=True, xticklabels=False, yticklabels=False)
    plt.xlabel("Posición X (bins)")
    plt.ylabel("Posición Y (bins)")
    plt.title(f"Diversidad Alfa Espacial en {sample}")
    plt.gca().invert_yaxis()
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")
    plt.show()

# --- EJEMPLO DE USO ---
csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"  # Ruta del archivo CSV
community_col = "Sample"  # Nombre de la columna que identifica la comunidad
species_col = "phenotype_key"  # Nombre de la columna que identifica las especies
#sample_to_analyze = "AGSCC_1"  # Comunidad a analizar
output_plot = f"simpson_heatmap_{sample_to_analyze}.png"

# Cargar datos
df = pd.read_csv(csv_file)

# Calcular diversidad espacial
diversity_results, x_bins, y_bins = calculate_spatial_diversity(df, community_col, species_col, sample_to_analyze, grid_size=200)

# Graficar la diversidad espacial como heatmap
plot_spatial_diversity(diversity_results, x_bins, y_bins, sample_to_analyze, output_plot)
```

### Índice de Diversidad de Shannon (H') - points

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def shannon_diversity_index(counts):
    """
    Calcula el índice de diversidad de Shannon (H').
    """
    total = sum(counts)
    if total == 0:
        return 0
    proportions = counts / total
    H = -sum(p * np.log(p) for p in proportions if p > 0)
    return H

def calculate_spatial_diversity(df, community_col, species_col, sample, grid_size=1000):
    """
    Calcula la diversidad alfa espacialmente dividiendo el área en una cuadrícula.
    """
    df_sample = df[df[community_col] == sample].copy()

    x_min, x_max = df_sample['X_centroid'].min(), df_sample['X_centroid'].max()
    y_min, y_max = df_sample['Y_centroid'].min(), df_sample['Y_centroid'].max()

    x_bins = np.arange(x_min, x_max, grid_size)
    y_bins = np.arange(y_min, y_max, grid_size)

    df_sample['X_bin'] = np.digitize(df_sample['X_centroid'], x_bins)
    df_sample['Y_bin'] = np.digitize(df_sample['Y_centroid'], y_bins)

    diversity_results = df_sample.groupby(['X_bin', 'Y_bin'])[species_col].apply(lambda x: shannon_diversity_index(x.value_counts())).reset_index()
    diversity_results.columns = ['X_bin', 'Y_bin', 'Shannon_Diversity_Index']

    diversity_results['X_coord'] = x_bins[diversity_results['X_bin'] - 1] + grid_size / 2
    diversity_results['Y_coord'] = y_bins[diversity_results['Y_bin'] - 1] + grid_size / 2

    return diversity_results

def plot_spatial_diversity(diversity_df, sample, output_plot="diversidad_espacial.png"):
    """
    Genera y guarda un mapa de calor de la diversidad alfa espacial dentro de una comunidad.
    """
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(diversity_df['X_coord'], diversity_df['Y_coord'], 
                          c=diversity_df['Shannon_Diversity_Index'], cmap='viridis', 
                          edgecolors='black', s=10, alpha=0.8)
    
    plt.colorbar(scatter, label="Índice de Diversidad de Shannon (H')")
    plt.xlabel("Posición X")
    plt.ylabel("Posición Y")
    plt.title(f"Diversidad Alfa Espacial en la Comunidad {sample}")
    
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")
    plt.show()

# --- EJEMPLO DE USO ---

csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"
#sample_to_analyze = "AGSCC_1"
output_plot = f"shannon_points_{sample_to_analyze}.png"

community_column = "Sample"  # Nombre de la columna que contiene la comunidad
species_column = "phenotype_key"  # Nombre de la columna con los nombres de especies

# Cargar datos
df = pd.read_csv(csv_file)

# Calcular diversidad espacial
diversity_results = calculate_spatial_diversity(df, community_column, species_column, sample_to_analyze, grid_size=600)

# Graficar la diversidad espacial
plot_spatial_diversity(diversity_results, sample_to_analyze, output_plot)
```

## Valor maximo de 1/D en funcion del grid_size

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def simpson_reciprocal_index(counts):
    """
    Calcula el índice de diversidad de Simpson recíproco (1/D).
    """
    total = sum(counts)
    if total == 0:
        return 0
    D = sum((n / total) ** 2 for n in counts)
    return 1 / D if D > 0 else 0

def calculate_spatial_diversity(df, community_col, species_col, sample, grid_size=100):
    """
    Calcula la diversidad alfa espacialmente dividiendo el área en una cuadrícula.
    """
    df_sample = df[df[community_col] == sample].copy()

    # Definir los límites del espacio
    x_min, x_max = df_sample['X_centroid'].min(), df_sample['X_centroid'].max()
    y_min, y_max = df_sample['Y_centroid'].min(), df_sample['Y_centroid'].max()

    # Crear la cuadrícula
    x_bins = np.arange(x_min, x_max, grid_size)
    y_bins = np.arange(y_min, y_max, grid_size)

    # Asignar cada punto a una celda de la cuadrícula
    df_sample['X_bin'] = np.digitize(df_sample['X_centroid'], x_bins)
    df_sample['Y_bin'] = np.digitize(df_sample['Y_centroid'], y_bins)

    # Agrupar por celda y calcular diversidad
    diversity_results = df_sample.groupby(['X_bin', 'Y_bin'])[species_col]\
                                 .apply(lambda x: simpson_reciprocal_index(x.value_counts()))\
                                 .reset_index()
    diversity_results.columns = ['X_bin', 'Y_bin', 'Simpson_Reciprocal_Index']

    return diversity_results, x_bins, y_bins

def plot_spatial_diversity(diversity_df, x_bins, y_bins, sample, output_plot="diversidad_espacial.png"):
    """
    Genera y guarda un mapa de calor de la diversidad alfa espacial dentro de una comunidad.
    """
    heatmap_data = np.full((len(y_bins), len(x_bins)), np.nan)

    for _, row in diversity_df.iterrows():
        # Ajuste para que los índices no se salgan de rango
        if (1 <= row['X_bin'] <= len(x_bins)) and (1 <= row['Y_bin'] <= len(y_bins)):
            heatmap_data[int(row['Y_bin']) - 1, int(row['X_bin']) - 1] = row['Simpson_Reciprocal_Index']

    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data, cmap='viridis', cbar=True, 
                xticklabels=False, yticklabels=False)
    plt.xlabel("Posición X (bins)")
    plt.ylabel("Posición Y (bins)")
    plt.title(f"Diversidad Alfa Espacial en {sample}")
    plt.gca().invert_yaxis()
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")
    plt.show()

def analyze_max_simpson_over_grids(df, community_col, species_col, sample, grid_sizes, 
                                   output_csv="max_simpson_values.csv", 
                                   output_plot="max_simpson_plot.png"):
    """
    Itera sobre diferentes valores de grid_size y calcula el valor máximo del 
    índice de Simpson recíproco en cada uno. Guarda los resultados en un .csv y 
    genera un gráfico mostrando la evolución de este valor máximo.
    """
    results = []

    for size in grid_sizes:
        diversity_results, x_bins, y_bins = calculate_spatial_diversity(df, community_col, species_col, 
                                                                        sample, grid_size=size)

        # Calcular el valor máximo del índice de Simpson recíproco
        max_val = diversity_results['Simpson_Reciprocal_Index'].max()
        
        results.append({
            'grid_size': size,
            'max_simpson_index': max_val if pd.notnull(max_val) else 0
        })

    # Convertir resultados a DataFrame y guardarlos en CSV
    df_results = pd.DataFrame(results)
    df_results.to_csv(output_csv, index=False)
    print(f"Resultados guardados en: {output_csv}")

    # Graficar la evolución del valor máximo
    plt.figure(figsize=(8, 6))
    plt.plot(df_results['grid_size'], df_results['max_simpson_index'], marker='o', color='blue')
    plt.xlabel('Tamaño de grilla (grid_size)')
    plt.ylabel('Máximo índice de Simpson recíproco')
    plt.title(f'Evolución del índice de Simpson recíproco máximo para {sample}')
    plt.grid(True)
    plt.savefig(output_plot, dpi=300, bbox_inches='tight')
    print(f"Gráfico guardado en: {output_plot}")
    plt.show()

# --- EJEMPLO DE USO ---
if __name__ == "__main__":
    # Rutas y columnas
    csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"
    community_col = "Sample"          # Columna que identifica la comunidad
    species_col = "phenotype_key"     # Columna que identifica la especie/fenotipo
    sample_to_analyze = "AGSCC_1"     # Muestra a analizar

    # Lista de tamaños de grilla que queremos probar
    grid_sizes_to_test = [10, 20, 30, 50, 60, 70, 80, 90, 100, 120, 140, 160, 180, 200]

    # Salidas
    output_csv_name = "max_simpson_values.csv"
    output_plot_name = "max_simpson_plot.png"

    # Cargar datos
    df_data = pd.read_csv(csv_file)

    # Ejecutar el análisis de varios grid_sizes
    analyze_max_simpson_over_grids(
        df=df_data,
        community_col=community_col,
        species_col=species_col,
        sample=sample_to_analyze,
        grid_sizes=grid_sizes_to_test,
        output_csv=output_csv_name,
        output_plot=output_plot_name
    )

    # Si además quisieras generar un heatmap de un grid_size en particular,
    # podrías llamar a las funciones existentes:
    #diversity_results, x_bins, y_bins = calculate_spatial_diversity(df_data, community_col, species_col, sample_to_analyze, grid_size=200)
    #plot_spatial_diversity(diversity_results, x_bins, y_bins, sample_to_analyze, "heatmap_200.png")
```

```python
# Parallel

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Para paralelizar
from joblib import Parallel, delayed

def simpson_reciprocal_index(counts):
    """
    Calcula el índice de diversidad de Simpson recíproco (1/D).
    """
    total = sum(counts)
    if total == 0:
        return 0
    D = sum((n / total) ** 2 for n in counts)
    return 1 / D if D > 0 else 0

def calculate_spatial_diversity(df, community_col, species_col, sample, grid_size=100):
    """
    Calcula la diversidad alfa espacialmente dividiendo el área en una cuadrícula.
    """
    df_sample = df[df[community_col] == sample].copy()

    # Definir los límites del espacio
    x_min, x_max = df_sample['X_centroid'].min(), df_sample['X_centroid'].max()
    y_min, y_max = df_sample['Y_centroid'].min(), df_sample['Y_centroid'].max()

    # Crear la cuadrícula
    x_bins = np.arange(x_min, x_max, grid_size)
    y_bins = np.arange(y_min, y_max, grid_size)

    # Asignar cada punto a una celda de la cuadrícula
    df_sample['X_bin'] = np.digitize(df_sample['X_centroid'], x_bins)
    df_sample['Y_bin'] = np.digitize(df_sample['Y_centroid'], y_bins)

    # Agrupar por celda y calcular diversidad
    diversity_results = (
        df_sample
        .groupby(['X_bin', 'Y_bin'])[species_col]
        .apply(lambda x: simpson_reciprocal_index(x.value_counts()))
        .reset_index()
    )
    diversity_results.columns = ['X_bin', 'Y_bin', 'Simpson_Reciprocal_Index']

    return diversity_results, x_bins, y_bins

def plot_spatial_diversity(diversity_df, x_bins, y_bins, sample, output_plot="diversidad_espacial.png"):
    """
    Genera y guarda un mapa de calor de la diversidad alfa espacial dentro de una comunidad.
    """
    heatmap_data = np.full((len(y_bins), len(x_bins)), np.nan)

    for _, row in diversity_df.iterrows():
        # Ajuste para que los índices no se salgan de rango
        if (1 <= row['X_bin'] <= len(x_bins)) and (1 <= row['Y_bin'] <= len(y_bins)):
            heatmap_data[int(row['Y_bin']) - 1, int(row['X_bin']) - 1] = row['Simpson_Reciprocal_Index']

    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data, cmap='viridis', cbar=True, xticklabels=False, yticklabels=False)
    plt.xlabel("Posición X (bins)")
    plt.ylabel("Posición Y (bins)")
    plt.title(f"Diversidad Alfa Espacial en {sample}")
    plt.gca().invert_yaxis()
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")
    plt.show()

def analyze_one_grid(df, community_col, species_col, sample, grid_size):
    """
    Calcula la diversidad para un grid_size dado y devuelve el valor máximo del 
    índice de Simpson recíproco.
    """
    diversity_results, _, _ = calculate_spatial_diversity(df, community_col, species_col, sample, grid_size)
    max_val = diversity_results['Simpson_Reciprocal_Index'].max()
    max_val = max_val if pd.notnull(max_val) else 0
    return {'grid_size': grid_size, 'max_simpson_index': max_val}

def analyze_max_simpson_over_grids(df, community_col, species_col, sample, grid_sizes, 
                                   n_jobs=-1,   # Por defecto, usa todos los núcleos disponibles
                                   output_csv="max_simpson_values.csv", 
                                   output_plot="max_simpson_plot.png"):
    """
    Itera sobre diferentes valores de grid_size (en paralelo, usando joblib) y 
    calcula el valor máximo del índice de Simpson recíproco para cada uno.
    Guarda los resultados en un .csv y genera un gráfico mostrando la evolución 
    de este valor máximo.
    
    Parámetros:
    -----------
    - df : DataFrame con los datos.
    - community_col : Columna que identifica la comunidad.
    - species_col : Columna que identifica la especie.
    - sample : Nombre de la muestra a analizar.
    - grid_sizes : Lista o rango de grid_sizes que se quieren probar.
    - n_jobs : Número de trabajos (procesos/hilos) en paralelo. 
               Si es -1, usa todos los núcleos.
    - output_csv : Nombre del archivo CSV de salida.
    - output_plot : Nombre del gráfico de salida.
    """

    # Paralelizar la ejecución de analyze_one_grid() para cada grid_size
    results = Parallel(n_jobs=n_jobs)(
        delayed(analyze_one_grid)(df, community_col, species_col, sample, size) 
        for size in grid_sizes
    )

    # Convertir resultados a DataFrame y guardarlos en CSV
    df_results = pd.DataFrame(results).sort_values(by='grid_size')
    df_results.to_csv(output_csv, index=False)
    print(f"Resultados guardados en: {output_csv}")

    # Graficar la evolución del valor máximo
    plt.figure(figsize=(8, 6))
    plt.plot(df_results['grid_size'], df_results['max_simpson_index'], 
             marker='o', color='blue')
    plt.xlabel('Tamaño de grilla (grid_size)')
    plt.ylabel('Máximo índice de Simpson recíproco')
    plt.title(f'Evolución del índice de Simpson recíproco máximo para {sample}')
    plt.grid(True)
    plt.savefig(output_plot, dpi=300, bbox_inches='tight')
    print(f"Gráfico guardado en: {output_plot}")
    plt.show()

# --- EJEMPLO DE USO ---
if __name__ == "__main__":
    # Rutas y columnas
    csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"
    community_col = "Sample"
    species_col = "phenotype_key"
    sample_to_analyze = "AGSCC_1"

    # Lista de tamaños de grilla a probar
    grid_sizes_to_test = [10, 20, 30, 50, 60, 70, 80, 90, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300]

    # Salidas
    output_csv_name = "max_simpson_values.csv"
    output_plot_name = "max_simpson_plot.png"

    # Cargar datos
    df_data = pd.read_csv(csv_file)

    # Ejecutar el análisis (paralelizado con joblib)
    analyze_max_simpson_over_grids(
        df=df_data,
        community_col=community_col,
        species_col=species_col,
        sample=sample_to_analyze,
        grid_sizes=grid_sizes_to_test,
        n_jobs=-1,  # -1 para usar todos los cores disponibles
        output_csv=output_csv_name,
        output_plot=output_plot_name
    )

    # Generar un heatmap para un grid_size específico
    # diversity_results, x_bins, y_bins = calculate_spatial_diversity(
    #     df_data, community_col, species_col, sample_to_analyze, grid_size=200)
    # plot_spatial_diversity(diversity_results, x_bins, y_bins, sample_to_analyze, "heatmap_200.png")
```

```python
# Shannon

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from joblib import Parallel, delayed

def shannon_index(counts):
    """
    Calcula el índice de diversidad de Shannon (H').
    """
    total = sum(counts)
    if total == 0:
        return 0
    # H' = -sum( (n_i / total) * ln(n_i / total) )
    return -sum((n / total) * np.log(n / total) for n in counts if n > 0)

def calculate_spatial_diversity(df, community_col, species_col, sample, grid_size=100):
    """
    Calcula la diversidad alfa (Shannon) espacialmente dividiendo el área en una cuadrícula.
    """
    df_sample = df[df[community_col] == sample].copy()

    # Definir los límites del espacio
    x_min, x_max = df_sample['X_centroid'].min(), df_sample['X_centroid'].max()
    y_min, y_max = df_sample['Y_centroid'].min(), df_sample['Y_centroid'].max()

    # Crear la cuadrícula
    x_bins = np.arange(x_min, x_max, grid_size)
    y_bins = np.arange(y_min, y_max, grid_size)

    # Asignar cada punto a una celda de la cuadrícula
    df_sample['X_bin'] = np.digitize(df_sample['X_centroid'], x_bins)
    df_sample['Y_bin'] = np.digitize(df_sample['Y_centroid'], y_bins)

    # Agrupar por celda y calcular diversidad (Shannon)
    diversity_results = (
        df_sample
        .groupby(['X_bin', 'Y_bin'])[species_col]
        .apply(lambda x: shannon_index(x.value_counts()))
        .reset_index()
    )
    diversity_results.columns = ['X_bin', 'Y_bin', 'Shannon_Index']

    return diversity_results, x_bins, y_bins

def plot_spatial_diversity(diversity_df, x_bins, y_bins, sample, output_plot="diversidad_espacial.png"):
    """
    Genera y guarda un mapa de calor de la diversidad alfa espacial (Shannon) dentro de una comunidad.
    """
    heatmap_data = np.full((len(y_bins), len(x_bins)), np.nan)

    for _, row in diversity_df.iterrows():
        # Ajuste para que los índices no se salgan de rango
        if (1 <= row['X_bin'] <= len(x_bins)) and (1 <= row['Y_bin'] <= len(y_bins)):
            heatmap_data[int(row['Y_bin']) - 1, int(row['X_bin']) - 1] = row['Shannon_Index']

    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data, cmap='viridis', cbar=True, xticklabels=False, yticklabels=False)
    plt.xlabel("Posición X (bins)")
    plt.ylabel("Posición Y (bins)")
    plt.title(f"Diversidad Alfa (Shannon) Espacial en {sample}")
    plt.gca().invert_yaxis()
    plt.savefig(output_plot, dpi=600, bbox_inches='tight')
    print(f"Gráfico guardado en {output_plot}")
    plt.show()

def calc_max_for_one_sample(df, community_col, species_col, sample, grid_size):
    """
    Calcula el valor máximo del índice de Shannon para una (muestra, grid_size).
    Devuelve un dict con sample, grid_size y max_shannon_index.
    """
    diversity_results, _, _ = calculate_spatial_diversity(
        df, community_col, species_col, sample, grid_size
    )
    max_val = diversity_results['Shannon_Index'].max()
    max_val = max_val if pd.notnull(max_val) else 0

    return {
        'sample': sample,
        'grid_size': grid_size,
        'max_shannon_index': max_val
    }

def compare_grid_sizes_across_samples(df, community_col, species_col, 
                                      grid_sizes, n_jobs=-1,
                                      output_csv="comparative_max_shannon_values.csv",
                                      output_plot="comparative_max_shannon_plot.png"):
    """
    Para cada muestra única en df[community_col], calcula (en paralelo) el valor
    máximo del índice de Shannon en cada grid_size.
    
    Genera un solo CSV con columnas [sample, grid_size, max_shannon_index],
    y un gráfico comparativo con la evolución del índice en función de grid_size
    para cada muestra.
    
    Parámetros:
    -----------
    - df : DataFrame con los datos.
    - community_col : str, nombre de la columna con la comunidad/muestra.
    - species_col : str, nombre de la columna con la especie/fenotipo.
    - grid_sizes : list de int, tamaños de grilla a probar.
    - n_jobs : int, número de núcleos/procesos para joblib. -1 usa todos.
    - output_csv : str, nombre del archivo CSV de salida.
    - output_plot : str, nombre del archivo PNG del gráfico comparativo.
    """

    # Obtiene todas las muestras únicas del DataFrame
    all_samples = df[community_col].unique()

    # Generamos la lista de tareas: (sample, grid_size) para cada muestra
    tasks = []
    for sample in all_samples:
        for gsize in grid_sizes:
            tasks.append((sample, gsize))

    # Ejecutamos en paralelo calc_max_for_one_sample
    results = Parallel(n_jobs=n_jobs)(
        delayed(calc_max_for_one_sample)(df, community_col, species_col, sample, gsize)
        for (sample, gsize) in tasks
    )

    # Convertir la lista de dicts en DataFrame
    df_results = pd.DataFrame(results)

    # Guardar resultados en CSV
    df_results.to_csv(output_csv, index=False)
    print(f"Resultados guardados en: {output_csv}")

    # Graficar: unimos cada muestra con su curva "grid_size vs. max_shannon_index"
    plt.figure(figsize=(10, 7))
    sns.lineplot(data=df_results, x='grid_size', y='max_shannon_index',
                 hue='sample', marker='o')

    plt.xlabel('Tamaño de grilla (grid_size)')
    plt.ylabel('Máximo índice de Shannon')
    plt.title('Comparativa de la evolución del índice de Shannon en distintas muestras')
    plt.grid(True)
    plt.legend(title='Muestra', bbox_to_anchor=(1.02, 1), loc='upper left')  # Leyenda fuera
    plt.tight_layout()
    plt.savefig(output_plot, dpi=300)
    print(f"Gráfico comparativo guardado en: {output_plot}")
    plt.show()

# --- EJEMPLO DE USO ---
if __name__ == "__main__":
    # Rutas y columnas
    csv_file = "/media/cruz/5129BE3F7A3D6E25/TMA/2_1_Phenotype_calling/aggregated_data.csv"
    community_col = "ROIs"          # Columna con la comunidad/muestra
    species_col = "phenotype_key"     # Columna con la especie/fenotipo

    # Valores de grid_size a analizar
    grid_sizes_to_test = [10, 20, 30, 50, 60, 70, 80, 90, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300]

    # Salidas
    output_csv_name = "comparative_max_shannon_values.csv"
    output_plot_name = "comparative_max_shannon_plot.png"

    # Cargar datos
    df_data = pd.read_csv(csv_file)

    # Ejecutar el análisis comparativo en paralelo
    compare_grid_sizes_across_samples(
        df=df_data,
        community_col=community_col,
        species_col=species_col,
        grid_sizes=grid_sizes_to_test,
        n_jobs=-1,  # Usa todos los cores disponibles
        output_csv=output_csv_name,
        output_plot=output_plot_name
    )

    # OPCIONAL: También puedes hacer un heatmap para una (muestra, grid_size) específica
    # diversity_results, x_bins, y_bins = calculate_spatial_diversity(
    #     df_data, community_col, species_col, "AGSCC_1", grid_size=200
    # )
    # plot_spatial_diversity(
    #     diversity_results, x_bins, y_bins, "AGSCC_1", 
    #     output_plot="heatmap_AGSCC_1_gs200.png"
    # )
```
