```python
import os
import shutil

# Path where the folders with the files of each cycle
base_path = "/media/cruz/Spatial/t-CycIF_human_2025_2/01_Registration/CYCLES/" # Cambia esta ruta según tu estructura

# Function to extract rcpnl files and put them in separate folders according to samples
def mover_archivo(cycle, sample_folder_name):
    # Obtener el nombre de la muestra antes del '@'
    sample_name = sample_folder_name.split('@')[0]
    
    # Ruta de la carpeta de la muestra dentro del ciclo
    sample_folder = os.path.join(base_path, cycle, sample_folder_name)
    
    # Buscar el archivo que termina en .rcpnl dentro de la carpeta de la muestra
    archivos = [f for f in os.listdir(sample_folder) if f.endswith('.rcpnl')]
    
    # Si encontramos exactamente un archivo .rcpnl, lo movemos
    if len(archivos) == 1:
        archivo = archivos[0]
        archivo_origen = os.path.join(sample_folder, archivo)
        
        # Crear la carpeta de destino si no existe
        destino = os.path.join(base_path, sample_name)
        if not os.path.exists(destino):
            os.makedirs(destino)
        
        # Ruta del archivo de destino (usamos el nombre del ciclo para diferenciar los archivos)
        destino_final = os.path.join(destino, f"{cycle}_{archivo}")
        
        # Mover el archivo
        shutil.move(archivo_origen, destino_final)
        print(f"Archivo {archivo} movido de {archivo_origen} a {destino_final}")
    else:
        print(f"Error: No se encontró o se encontró más de un archivo .rcpnl en {sample_folder}")

# Recorrer los ciclos (CYCLE 1 a CYCLE 6)
for i in range(1, 7):
    cycle = f"CYCLE {i}"
    
    # Recorrer las carpetas de muestras dentro de cada ciclo
    for sample_folder_name in os.listdir(os.path.join(base_path, cycle)):
        # Asegurarse de que estamos trabajando solo con las carpetas de muestra
        if os.path.isdir(os.path.join(base_path, cycle, sample_folder_name)):
            # Mover el archivo de la muestra en este ciclo
            mover_archivo(cycle, sample_folder_name)
```

```python
import os
import shutil

# Ruta base donde están las carpetas CYCLE 1, CYCLE 2, etc.
base_path = "/media/cruz/Spatial/t-CycIF_human_2025_2/01_Registration/CYCLES"

def mover_archivo(cycle_folder, archivo):
    # Extraer la muestra: parte antes del '@' en el nombre del archivo
    muestra = archivo.split('@')[0].strip()  # Ej: "3_F13P1_1_3"
    
    # Crear carpeta de destino para la muestra
    destino_muestra = os.path.join(base_path, "..", "Muestras_Organizadas", muestra)
    os.makedirs(destino_muestra, exist_ok=True)
    
    # Nuevo nombre del archivo (incluye el ciclo)
    nombre_final = f"{os.path.basename(cycle_folder)}_{archivo}"
    
    # Mover el archivo
    shutil.move(
        os.path.join(cycle_folder, archivo),
        os.path.join(destino_muestra, nombre_final)
    )
    print(f"Archivo {archivo} movido a {destino_muestra}/{nombre_final}")

# Procesar cada ciclo (CYCLE 1 a CYCLE 7)
for cycle_num in range(1, 8):
    cycle_folder = os.path.join(base_path, f"CYCLE {cycle_num}")
    
    # Procesar todos los archivos .rcpnl en la carpeta del ciclo
    for archivo in os.listdir(cycle_folder):
        if archivo.endswith(".rcpnl"):
            mover_archivo(cycle_folder, archivo)
```
